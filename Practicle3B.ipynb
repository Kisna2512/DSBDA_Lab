```python
import pandas as pd
import numpy as np
data=pd.read_csv("Iris.csv")

print(data.describe())
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
print(df.isnull())
```

```python
setosa = df[~df["Species"].isin(['Iris-virginica','Iris-versicolor'])]
print(setosa)
```

```python
virginica = df[~df["Species"].isin(['Iris-setosa','Iris-versicolor'])]
virginica
```
